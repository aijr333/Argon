```python
# Run this cell first
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, sin, cos, simplify, init_printing
from scipy.fft import fft
from ipywidgets import interact, FloatSlider
from IPython.display import display
init_printing()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
sine = np.sin(x)
cosine = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, sine, label='sin(x)')
plt.plot(x, cosine, label='cos(x)', linestyle='--')
plt.title('Sine and Cosine Waves')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

# Shows cyclical behavior — good for anything that repeats!
# Formula: y = A * sin(Bx + C)

# USES:
# - Science: Describes waves, light, sound
# - Business: Predict seasonality or customer rhythm
# - Engineering: Models oscillations, AC currents
# - Physics: Simulates pendulums, circular motion

# MANIPULATE:
# A = amplitude → how tall the wave is  
# B = frequency → how fast it repeats  
# C = phase → horizontal shift (start position)


```python
x = symbols('x')
f = x**2 * sin(x)
df = diff(f, x)
print("Function:")
display(f)
print("Derivative:")
display(simplify(df))
```

# Finds the rate of change — "how fast something grows or shrinks"
# Formula: d/dx [f(x)]

# USES:
# - Business: Marginal profit/cost (change per unit)
# - Science: Population growth models
# - Physics: Speed, acceleration, energy rate
# - AI: Gradient descent in training models

# MANIPULATE:
# Change f(x) to any function — test impact of growth
# e.g. x**3 * sin(x), log(x), e^x


```python
def projectile(v=20, theta_deg=45):
    g = 9.81
    theta = np.radians(theta_deg)
    t_flight = (2 * v * np.sin(theta)) / g
    t = np.linspace(0, t_flight, num=200)

    x = v * np.cos(theta) * t
    y = v * np.sin(theta) * t - 0.5 * g * t**2

    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    plt.title(f'Projectile Motion (v={v} m/s, θ={theta_deg}°)')
    plt.xlabel('Distance (m)')
    plt.ylabel('Height (m)')
    plt.grid(True)
    plt.show()

projectile()
```

# Plots a parabolic arc — models motion with gravity
# Formula: y = x*tan(θ) - (g * x^2) / (2 * v^2 * cos²(θ))

# MANIPULATE:

```python
# Signal example: square wave
T = 2*np.pi
x = np.linspace(0, T, 1000)
f = np.sign(np.sin(x))

# Fourier transform
F = fft(f)
frequencies = np.abs(F[:len(F)//2])

plt.figure(figsize=(10, 5))
plt.plot(frequencies[:50])
plt.title('Fourier Transform of Square Wave')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

# Converts a signal into its frequencies
# Formula: FFT(f(x)) → gives magnitudes of sin/cos parts

# USES:
# - Engineering: Signal processing, sensors, vibration analysis
# - Audio: Noise removal, compression (MP3)
# - Business: Detect patterns (e.g., sales cycles)
# - Science: Analyze waveforms, brainwaves

# MANIPULATE:
# Input your own data (sales, heartbeats, etc.)
# Use FFT to find dominant cycles or spikes


```python
def plot_sine(A=1, f=1, phi=0):
    x = np.linspace(0, 2 * np.pi, 500)
    y = A * np.sin(2 * np.pi * f * x + phi)

    plt.figure(figsize=(10, 4))
    plt.plot(x, y)
    plt.title(f'Sine Wave: A={A}, f={f}, φ={phi}')
    plt.xlabel('x')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.ylim(-2, 2)
    plt.show()

interact(plot_sine,
         A=FloatSlider(value=1, min=0, max=2, step=0.1),
         f=FloatSlider(value=1, min=0.5, max=5, step=0.1),
         phi=FloatSlider(value=0, min=0, max=2*np.pi, step=0.1))
```
